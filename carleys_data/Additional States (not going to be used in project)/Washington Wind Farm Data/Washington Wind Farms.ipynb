```python
import requests
import json
import pandas as pd
```

```python
url = "https://eersc.usgs.gov/api/uswtdb/v1/turbines?&t_state=eq.WA"
```

```python
print(requests.get(url))
```

```python
response = requests.get(url).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
clean_wa_df = washington_df[["t_county", "p_name", "p_year", "p_tnum", "p_cap", "xlong", "ylat"]]
clean_wa_df
```

```python
wa_df = clean_wa_df.rename(columns={"t_county": "County", 
                                    "p_name": "Wind Farm Name", 
                                    "p_year": "Year", 
                                    "p_tnum": "Turbine Count", 
                                    "p_cap": "Cumulative Capacity of Wind Farm (in MW)", 
                                    "xlong": "Longitude", 
                                    "ylat": "Latitude"
                                   })
wa_df
```

```python
wa_df.dropna(subset=["Year"], inplace=True)
wa_df
```

```python
current_wa_df = wa_df[wa_df["Year"] >= 2000]
current_wa_df
```
